```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('temp.csv')
df.head()
```

```python
temp_df = df[ ['Year', 'Statistic', 'Month', 'Value'] ]
temp_df = temp_df[(temp_df['Statistic'] == 'Mean') & (temp_df['Year'] >= 2557)]
temp_df
```

```python
temp_df.info()
```

```python
temp_df['Value'] = temp_df['Value'].astype('float')
```

### Data from 3 station
- calculate mean for each month and year

```python
temp_df = temp_df.groupby(['Year', 'Month']).agg({'Value': 'mean'}).reset_index()
temp_df['Year'] = temp_df['Year'] - 543
temp_df.head()
```

```python
month_dict = {'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr': 4, 'May': 5, 'Jun': 6, 
              'Jul': 7, 'Aug': 8, 'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec': 12}
temp_df['Month'] = temp_df['Month'].map(month_dict)
temp_df.sort_values(['Year', 'Month'])
```

## Prepare Year 2023, 2024

```python
temp_cat = pd.read_csv('tmax-tmin.csv')
temp_cat
```

```python
temp_cat = temp_cat[temp_cat['ProvinceID'] == 50]
temp_cat = temp_cat.drop(columns=['ProvinceID', 'ProvinceName'])
temp_cat.head()
```

```python
temp_cat.info()
```

```python
temp_cat['Tmax'] = temp_cat['Tmax'].astype('float')
temp_cat['Tmin'] = temp_cat['Tmin'].astype('float')
```

```python
temp_cat['Value'] = (temp_cat['Tmax'] + temp_cat['Tmin']) / 2
temp_cat
```

```python
temp_cat = temp_cat[temp_cat['Year'] >= 2023]
temp_cat = temp_cat.drop(columns=['Tmax', 'Tmin'])
```

```python
cat_prep = temp_cat.reset_index(drop=True)
cat_prep
```

## Final

```python
temp_final = pd.concat([temp_df, cat_prep[:18]])
temp_final = temp_final.sort_values(['Year', 'Month']).reset_index(drop=True)
temp_final
```

```python
temp_final.to_csv('temp_cleaned.csv')
```
