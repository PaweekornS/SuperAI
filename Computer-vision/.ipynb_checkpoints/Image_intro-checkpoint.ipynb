### Thai Flag

```python
Thai = np.ones((6, 9, 3))   # open figure with all white color
Thai[[0, -1], :, 1:] = 0  # red -> 1: close GB
Thai[2:4, :, :2] = 0  # blue  -> :2 close RG
plt.imshow(Thai)
plt.axis('off')
```

```python
import cv2

# wrong color because default color of opencv is BGR
cv2.imshow('Thai', cv2.resize(Thai, None, fx=100, fy=100))  # interpolation
cv2.waitKey()
```

### Japan Flag

```python
Japan = np.ones((600, 900, 3))
# proportional of red circle with figure
r = (3/5)/2 * Japan.shape[0]  # 3/5 * height = diameter

c = [i//2 for i in Japan.shape[:2]]  # center point

for i in range(Japan.shape[0]):
    for j in range(Japan.shape[1]):
        # circle formula: (y-k)^2 + (x-h)^2 = r^2
        if (i-c[0])**2 + (j-c[1])**2 <= r**2:
            Japan[i, j, 1:] = 0
            
plt.imshow(Japan)
```

## Import an Image

1. Matplotlib
- can only read PNGs natively. Further image formats are supported via the optional dependency on Pillow.
- <u>Pro</u>: get RGB for sure
- <u>Con</u>: Slow

```python
im = plt.imread('parrot.jpg')
plt.imshow(im)
```

2. Pillow
- Python Image Library

```python
from PIL import Image

im = Image.open('parrot.jpg')
im = np.array(im)
plt.imshow(im)
```

3. OpenCV
- <u>Pro</u>: Fast and various optimization
- <u>Con</u>: need to use their own library for display

```python
im = cv2.imread('parrot.jpg')
cv2.imshow('im', im)
cv2.waitKey()
```

### 2. Intensity (gray) level
- bit per channel
-  bit  visualization 

```python
def image2bitplane(img):
    bp = []
    for c in range(img.shape[2]):
        for i in range(8):
            bp.append(img[:, :, c] // 2 ** i % 2)            
    return np.array(bp)


# convert pixel into bits
def bitplane2image(bp):
    img = np.zeros( (bp[0].shape[0], bp[0].shape[1], 3) )
    for b in range(len(bp)):
        img[:, :, b//8] = img[:, :, b//8] + (bp[b] * 2**(b%8))
    return np.uint8(img)
```

```python
for bpc in range(1, 9):
    bp = image2bitplane(im)
    # replace the missing bit with 0
    bp[0:8-bpc, :, :] = 0  # R
    bp[8:8+8-bpc, :, :] = 0  # G
    bp[16:16+8-bpc, :, :] = 0  # B
    
    # convert back to picture
    im_ = bitplane2image(bp)
    max_vl = sum([2**i for i in range(8-bpc, 8)])
    im_ = np.uint8(im_ / max_vl * 255)
    plt.subplot(2, 4, bpc)
    plt.imshow(im_)
    plt.axis('off')
    plt.title(str(bpc))
    
plt.show()
```

## Color Detection
- RGB color space\
<u>Note</u>: OpenCV will be BGR

```python
A = cv2.imread('parrot.jpg')

tol = 30
color = np.array([64, 116, 118])  # BGR format

mask = cv2.inRange(A, color-tol, color+tol)
A_mask = cv2.bitwise_and(A, A, mask=mask)  # and bit with A
cv2.imshow('mask', mask)
cv2.imshow('A_mask', A_mask)
cv2.waitKey()
cv2.imwrite('mask.png', mask)
cv2.imwrite('A_mask.png', A_mask)
```
